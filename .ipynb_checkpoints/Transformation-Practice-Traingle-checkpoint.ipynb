```python
import numpy as np
import matplotlib.pyplot as plt

def scaling(p, sx, sy):
    # Define the scaling transformation matrix
    s = np.array([[sx, 0],  # Scale by sx in the x direction
                  [0, sy]]) # Scale by sy in the y direction
    # Apply the scaling transformation to each point
    r = np.dot(p, s)
    return r

def plot_func(original, scaled, title):
    plt.figure()
    # Plot original triangle
    plt.fill(*zip(*original), alpha=0.5, label='Original Triangle', color='blue')
    # Plot scaled triangle
    plt.fill(*zip(*scaled), alpha=0.5, label='Scaled Triangle', color='red')
    plt.title(title)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.axhline(0, color='black', linewidth=0.5, ls='--')
    plt.axvline(0, color='black', linewidth=0.5, ls='--')
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.legend()
    plt.axis('equal')
    plt.show()

# Define the vertices of the triangle
p = np.array([[2, 4],   # Vertex A
              [6, 4],   # Vertex B
              [4, 8]])  # Vertex C

# Scale factors
sx, sy = 2, 2

# Perform scaling
r = scaling(p, sx, sy)

# Plot the original and scaled triangles
plot_func(p, r, "2D Scaling of a Triangle")
```

# Rotation 45 Degree

```python
import numpy as np
import matplotlib.pyplot as plt

def rotation(p, angle):
    # Convert angle from degrees to radians
    theta = np.radians(angle)
    # Define the rotation transformation matrix
    r = np.array([[np.cos(theta), -np.sin(theta)], 
                  [np.sin(theta), np.cos(theta)]])
    # Apply the rotation transformation to each point
    r_transformed = np.dot(p, r)
    return r_transformed

def plot_func(original, rotated, title):
    plt.figure()
    # Plot original triangle
    plt.fill(*zip(*original), alpha=0.5, label='Original Triangle', color='blue')
    # Plot rotated triangle
    plt.fill(*zip(*rotated), alpha=0.5, label='Rotated Triangle', color='red')
    plt.title(title)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.axhline(0, color='black', linewidth=0.5, ls='--')
    plt.axvline(0, color='black', linewidth=0.5, ls='--')
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.legend()
    plt.axis('equal')
    plt.show()

# Define the vertices of the triangle
p = np.array([[2, 4],   # Vertex A
              [6, 4],   # Vertex B
              [4, 8]])  # Vertex C

# Perform 45-degree rotation
rotated_points = rotation(p, 45)

# Plot the original triangle and the rotated triangle
plot_func(p, rotated_points, "2D 45-Degree Rotation of a Triangle")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def translation(p, tx, ty):
    # Create the translation vector
    t = np.array([tx, ty])
    # Translate each point by adding the translation vector
    translated_points = p + t
    return translated_points

def plot_func(original, translated, title):
    plt.figure()
    # Plot original triangle
    plt.fill(*zip(*original), alpha=0.5, label='Original Triangle', color='blue')
    # Plot translated triangle
    plt.fill(*zip(*translated), alpha=0.5, label='Translated Triangle', color='red')
    plt.title(title)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.axhline(0, color='black', linewidth=0.5, ls='--')
    plt.axvline(0, color='black', linewidth=0.5, ls='--')
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.legend()
    plt.axis('equal')
    plt.show()

# Define the vertices of the triangle
p = np.array([[2, 4],   # Vertex A
              [6, 4],   # Vertex B
              [4, 8]])  # Vertex C

# Translation values
tx, ty = 3, 2  # Translate by 3 units in x and 2 units in y

# Perform translation
translated_points = translation(p, tx, ty)

# Plot the original triangle and the translated triangle
plot_func(p, translated_points, "2D Translation of a Triangle")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def reflection(p):
    # Define the reflection transformation matrix for the x-axis
    r = np.array([[1, 0], 
                  [0, -1]])
    # Apply the reflection transformation to each point
    reflected_points = np.dot(p, r)
    return reflected_points

def plot_func(original, reflected, title):
    plt.figure()
    # Plot original triangle
    plt.fill(*zip(*original), alpha=0.5, label='Original Triangle', color='blue')
    # Plot reflected triangle
    plt.fill(*zip(*reflected), alpha=0.5, label='Reflected Triangle', color='red')
    plt.title(title)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.axhline(0, color='black', linewidth=0.5, ls='--')
    plt.axvline(0, color='black', linewidth=0.5, ls='--')
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.legend()
    plt.axis('equal')
    plt.show()

# Define the vertices of the triangle
p = np.array([[2, 4],   # Vertex A
              [6, 4],   # Vertex B
              [4, 8]])  # Vertex C

# Perform reflection across the x-axis
reflected_points = reflection(p)

# Plot the original triangle and the reflected triangle
plot_func(p, reflected_points, "Reflection of a Triangle Across the X-axis")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def reflect_point(x, y, m, c):
    # Compute the reflected coordinates based on the line y = mx + c
    x_reflected = x - (2 * m * (m * x - y + c)) / (1 + m**2)
    y_reflected = y + (2 * (m * x - y + c)) / (1 + m**2)
    return x_reflected, y_reflected

def reflect_triangle(p, m, c):
    # Reflect each point of the triangle
    reflected_points = np.array([reflect_point(x, y, m, c) for x, y in p])
    return reflected_points

def plot_func(original, reflected, title, m, c):
    plt.figure()
    # Plot original triangle
    plt.fill(*zip(*original), alpha=0.5, label='Original Triangle', color='blue')
    # Plot reflected triangle
    plt.fill(*zip(*reflected), alpha=0.5, label='Reflected Triangle', color='red')
    
    # Plot the line y = mx + c
    x_vals = np.linspace(-1, 10, 200)
    y_vals = m * x_vals + c
    plt.plot(x_vals, y_vals, color='green', label=f'Line y = {m}x + {c}')

    plt.title(title)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.axhline(0, color='black', linewidth=0.5, ls='--')
    plt.axvline(0, color='black', linewidth=0.5, ls='--')
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.legend()
    plt.axis('equal')
    plt.show()

# Define the vertices of the triangle
p = np.array([[2, 4],   # Vertex A
              [6, 4],   # Vertex B
              [4, 8]])  # Vertex C

# Define the line parameters (slope m and y-intercept c)
m = 1  # Example slope
c = 1  # Example y-intercept

# Perform reflection across the line y = mx + c
reflected_points = reflect_triangle(p, m, c)

# Plot the original triangle and the reflected triangle
plot_func(p, reflected_points, "Reflection of a Triangle Across the Line y = mx + c", m, c)
```
