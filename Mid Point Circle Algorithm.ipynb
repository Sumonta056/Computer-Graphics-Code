```python
import matplotlib.pyplot as plt

# Function to set pixel (for visualization, we'll just store the points)
def set_pixel(x, y, points):
    points.append((x, y))

# Function to draw circle symmetry points
def draw_circle(xc, yc, x, y, points):
    set_pixel(xc + x, yc + y, points)
    set_pixel(xc - x, yc + y, points)
    set_pixel(xc + x, yc - y, points)
    set_pixel(xc - x, yc - y, points)
    set_pixel(xc + y, yc + x, points)
    set_pixel(xc - y, yc + x, points)
    set_pixel(xc + y, yc - x, points)
    set_pixel(xc - y, yc - x, points)

# Midpoint Circle Algorithm
def midpoint_circle(x_center, y_center, radius):
    x = 0
    y = radius
    d = 1 - radius  # Initial decision parameter
    points = []  # To store the points for plotting
    draw_circle(x_center, y_center, x, y, points)

    while x < y:
        x += 1
        if d < 0:
            d = d + 2 * x + 1  # Move to the next pixel in x direction
        else:
            y -= 1
            d = d + 2 * (x - y) + 1  # Move to the next pixel in both directions
        draw_circle(x_center, y_center, x, y, points)

    return points

# Example usage
x_center = 0
y_center = 0
radius = 10
circle_points = midpoint_circle(x_center, y_center, radius)

# Plotting the circle
plt.figure(figsize=(6, 6))
x_values, y_values = zip(*circle_points)
plt.scatter(x_values, y_values, color='blue', s=10)
plt.title("Midpoint Circle Algorithm")
plt.xlim(-radius - 1, radius + 1)
plt.ylim(-radius - 1, radius + 1)
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.show()
```
