```python
import matplotlib.pyplot as plt

# Function to set pixel (for visualization, we'll just store the points)
def set_pixel(canvas, x, y):
    canvas.append((x, y))

# Function to set label (for demonstration purposes, we'll just print the label)
def set_label(x, y, label):
    print(label)

# DDA Line Algorithm function
def dda_line(canvas, x0, y0, x1, y1):
    set_label(x0, y0, f"ddaLine({x0}, {y0}, {x1}, {y1})")

    # Calculate the differences
    dx = x1 - x0
    dy = y1 - y0

    # Calculate the number of steps needed
    steps = int(max(abs(dx), abs(dy)))

    # Calculate the increment in x and y for each step
    x_increment = dx / steps
    y_increment = dy / steps

    x, y = x0, y0
    for _ in range(steps + 1):
        set_pixel(canvas, round(x), round(y))  # Round to get the nearest pixel
        x += x_increment
        y += y_increment

# Example usage
canvas_points = []

# Draw a line from (1, 2) to (10, 5)
dda_line(canvas_points, 1, 2, 10, 5)

# Plotting the line
plt.figure(figsize=(6, 6))
x_values, y_values = zip(*canvas_points)
plt.scatter(x_values, y_values, color='red', s=10)
plt.plot(x_values, y_values, color='blue')  # Draw the line
plt.title("DDA Line Algorithm")
plt.xlim(0, 12)
plt.ylim(0, 8)
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.show()
```
